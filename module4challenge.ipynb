```python
#impor libraries for project 
import pandas as pd
import numpy as np 
from pathlib import Path
%matplotlib inline 
```

```python
#import document needed
whale_navs = pd.read_csv(
    Path('c:/users/mensg/fintech/mod4/mod4challenge/whale_navs.csv'), 
    index_col='date', 
    parse_dates=True, 
    infer_datetime_format=True
)
```

```python
#print head (top 5)
whale_navs.head()
```

```python
#daily returns = pct change 
#dropna so there wont be any empty data 
daily_returns = whale_navs.pct_change().dropna()
```

```python
#print head
daily_returns.head()
```

```python
#plot to compare
daily_returns.plot(figsize=(15,7), title='daily returns')
```

```python
#cumulative returns formula
cumulative_returns = (1 + daily_returns).cumprod()
```

```python
#cumulative returns tail
cumulative_returns.tail()
```

```python
#plot cumulative returns
cumulative_returns.plot(figsize=(15,15), title = 'cumulative returns')
```

```python
#annual std
annual_standard_deviation = daily_returns.std() * np.sqrt(trading_days)
```

```python
#1 yr returns
annual_average_returns = daily_returns.mean() * trading_days
```

```python
#sort
annual_average_returns.sort_values()
```

```python
#sharpe ratios
sharpe_ratios = annual_average_returns / annual_standard_deviation
```

```python
#sort values
sharpe_ratios.sort_values()
```

```python
#plot
sharpe_ratios.plot(figsize = (15,7), kind = 'bar', title = 'sharpe ratio')
```

```python
#market variance 
market_variance = daily_returns['S&P 500'].rolling(window=60).var()
```

```python
#tail
daily_returns['S&P 500'].rolling(window=60).var().tail()
```

```python
#brk covariance and tail
brk_covariance = daily_returns['BERKSHIRE HATHAWAY INC'].rolling(window=60).cov(daily_returns['S&P 500'])
brk_covariance.tail()
```

```python
#soros covariance and tail
soros_covariance = daily_returns['SOROS FUND MANAGEMENT LLC'].rolling(window=60).cov(daily_returns['S&P 500'])
brk_covariance.tail()
```

```python
#brk beta
brk_beta_60 = brk_covariance / market_variance 
```

```python
#print
brk_beta_60
```

```python
#soros beta
soros_beta_60 = soros_covariance / market_variance
```

```python
#print
soros_beta_60.tail()
```

```python
brk_beta_60.plot(figsize = (15,7), title = 'beta')
```

```python
soros_beta_60.plot(figsize = (15,7), title = 'beta')
```
